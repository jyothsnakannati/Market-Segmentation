```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

# Set seed for reproducibility
np.random.seed(1234)

# Range of clusters
n_clusters_range = range(2, 9)

# Store models and metrics
models = {}
inertia = []
silhouette_avg = []

for n_clusters in n_clusters_range:
    kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=1234)
    kmeans.fit(MD_x)
    
    # Store model
    models[n_clusters] = kmeans
    
    # Compute inertia (sum of squared distances to nearest cluster center)
    inertia.append(kmeans.inertia_)
    
    # Compute silhouette score
    if len(np.unique(kmeans.labels_)) > 1:
        silhouette_avg.append(silhouette_score(MD_x, kmeans.labels_))
    else:
        silhouette_avg.append(-1)  # Assign -1 if there's only one cluster

# Print results
print("Inertia for each number of clusters:", inertia)
print("Silhouette score for each number of clusters:", silhouette_avg)
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.utils import resample
from sklearn.metrics import adjusted_rand_score

# Set seed for reproducibility
np.random.seed(1234)

# Parameters
n_clusters_range = range(2, 9)
n_bootstrap = 100
n_repeats = 10

# Store results
bootstrap_results = {n: [] for n in n_clusters_range}

# Perform clustering for each number of clusters
for n_clusters in n_clusters_range:
    # Store bootstrap scores for each number of clusters
    cluster_labels = []
    for _ in range(n_repeats):
        # Perform clustering on original data
        kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=1234)
        kmeans.fit(MD_x)
        labels = kmeans.labels_
        cluster_labels.append(labels)
    
    # Perform bootstrapping
    for _ in range(n_bootstrap):
        # Generate bootstrap sample
        MD_x_bootstrap = resample(MD_x, random_state=1234)
        for labels in cluster_labels:
            # Fit k-means on the bootstrap sample
            kmeans_bootstrap = KMeans(n_clusters=n_clusters, n_init=10, random_state=1234)
            kmeans_bootstrap.fit(MD_x_bootstrap)
            # Compute stability using adjusted Rand index
            ari = adjusted_rand_score(labels, kmeans_bootstrap.labels_)
            bootstrap_results[n_clusters].append(ari)

# Convert results to mean and standard deviation for each number of clusters
mean_ari = {n: np.mean(bootstrap_results[n]) for n in n_clusters_range}
std_ari = {n: np.std(bootstrap_results[n]) for n in n_clusters_range}

# Print results
print("Mean Adjusted Rand Index for each number of clusters:")
print(mean_ari)
print("\nStandard Deviation of Adjusted Rand Index for each number of clusters:")
print(std_ari)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler

# Example DataFrame
mcdonalds = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9],
    'D': [10, 11, 12],
    'E': [13, 14, 15],
    'F': [16, 17, 18],
    'G': [19, 20, 21],
    'H': [22, 23, 24],
    'I': [25, 26, 27],
    'J': [28, 29, 30],
    'K': [31, 32, 33]
})

mcdonalds['Like.n'] = [1, 2, 3]  # Replace with actual target variable

X = mcdonalds.drop(columns=['Like.n'])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fit the GMM
gmm_refit = GaussianMixture(n_components=2, n_init=10, random_state=1234)
gmm_refit.fit(X_scaled)
# Predict the cluster labels
labels = gmm_refit.predict(X_scaled)

# Plotting
plt.figure(figsize=(10, 6))

# Scatter plot of the data points
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k')
plt.title('GMM Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Plot the GMM components (means and covariances)
ax = plt.gca()
for i in range(gmm_refit.n_components):
    mean = gmm_refit.means_[i]
    covar = gmm_refit.covariances_[i]

    # Check if covariance matrix is 2D
    if covar.shape[0] == 2 and covar.shape[1] == 2:
        # Calculate the ellipse properties
        v, w = np.linalg.eigh(covar)
        u = w[:, np.argmax(v)]
        v = np.sqrt(v[np.argmax(v)])
        angle = np.arctan2(u[1], u[0])
        angle = 180.0 * angle / np.pi

        # The Ellipse needs the width and height to be positive
        v = 2.0 * v

        # Create the ellipse
        ell = Ellipse(mean, v[0], v[1], 180.0 + angle, color='red', alpha=0.3)
        ax.add_patch(ell)

plt.colorbar(label='Cluster')
plt.show()
```

```python
# Example DataFrame
mcdonalds = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Male', 'Female'],
    'Cluster': [1, 2, 1, 2]  # Example cluster assignments
})
# Create a binary variable indicating whether each record is Female
mcdonalds['Is_Female'] = (mcdonalds['Gender'] == 'Female').astype(int)

# Group by cluster and calculate the mean of the binary variable
female_proportion = mcdonalds.groupby('Cluster')['Is_Female'].mean()

# Print the result
print(female_proportion)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data
visit = pd.Series([2.5, 3.0, 3.5, 4.0], index=[1, 2, 3, 4])
like = pd.Series([1.5, 2.0, 2.5, 3.0], index=[1, 2, 3, 4])
female_proportion = pd.Series([0.5, 0.7, 0.6, 0.8], index=[1, 2, 3, 4])
plt.figure(figsize=(10, 6))

# Scatter plot
plt.scatter(visit, like, s=10 * female_proportion * 100, alpha=0.6, edgecolors='w', linewidth=0.5)

# Add labels
for i in visit.index:
    plt.text(visit.loc[i], like.loc[i], str(i), fontsize=12, ha='center', va='center')

# Set axis limits
plt.xlim(2, 4.5)
plt.ylim(-3, 3)

# Label axes
plt.xlabel('Visit Frequency')
plt.ylabel('Like.n')
plt.title('Visit Frequency vs Like.n with Proportion of Females')

# Show the plot
plt.grid(True)
plt.show()
```
